```python
import time
import psutil
import matplotlib.pyplot as plt
import sys
sys.setrecursionlimit(20000)  # Augmenter la limite

# Fonctions des algorithmes de palindrome

def palindrome_reverse(chaine):
    return chaine == chaine[::-1]

def palindrome_two_pointers(chaine):
    gauche, droite = 0, len(chaine) - 1
    while gauche < droite:
        if chaine[gauche] != chaine[droite]:
            return False
        gauche += 1
        droite -= 1
    return True

def palindrome_recursion(chaine):
    if len(chaine) <= 1:
        return True
    elif chaine[0] != chaine[-1]:
        return False
    else:
        return palindrome_recursion(chaine[1:-1])

def palindrome_stack(chaine):
    pile = []
    n = len(chaine)
    for i in range(n // 2):
        pile.append(chaine[i])
    for i in range((n + 1) // 2, n):
        if pile.pop() != chaine[i]:
            return False
    return True

# Mesure du temps d'exécution et de la consommation de mémoire

def mesure_performance(fonction, chaine):
    # Mesurer la consommation de mémoire avant
    mem_avant = psutil.Process().memory_info().rss / 1024 ** 2  # en MB

    # Mesurer le temps de départ
    debut = time.time()

    # Appeler l'algorithme
    fonction(chaine)

    # Mesurer le temps après l'exécution
    fin = time.time()

    # Mesurer la consommation de mémoire après
    mem_apres = psutil.Process().memory_info().rss / 1024 ** 2  # en MB

    # Temps d'exécution et consommation de mémoire
    temps_execution = fin - debut
    memoire_utilisee = mem_apres - mem_avant

    return temps_execution, memoire_utilisee

# Tester les algorithmes avec une chaîne relativement longue
chaine_test = "a" * 10000 + "b" + "a" * 10000  # Chaîne non-palindrome pour test

# Algorithmes à tester
algorithmes = [
    ("Renversement de chaîne", palindrome_reverse),
    ("Pointeurs aux extrémités", palindrome_two_pointers),
    ("Récursion", palindrome_recursion),
    ("Pile", palindrome_stack)
]

# Résultats
temps_execution = []
memoire_utilisee = []

# Mesurer les performances pour chaque algorithme
for nom, algo in algorithmes:
    temps, memoire = mesure_performance(algo, chaine_test)
    temps_execution.append(temps)
    memoire_utilisee.append(memoire)

# Étape 3 : Créer un graphique comparatif

plt.figure(figsize=(10, 5))

# Graphe du temps d'exécution
plt.subplot(1, 2, 1)
plt.barh([nom for nom, _ in algorithmes], temps_execution, color='blue')
plt.xlabel('Temps d\'exécution (secondes)')
plt.title('Comparaison du temps d\'exécution')

# Graphe de la consommation mémoire
plt.subplot(1, 2, 2)
plt.barh([nom for nom, _ in algorithmes], memoire_utilisee, color='green')
plt.xlabel('Mémoire utilisée (MB)')
plt.title('Comparaison de la consommation de mémoire')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def mesure_temps_execution(fonction, chaine):
    debut = time.time()
    fonction(chaine)
    fin = time.time()
    return fin - debut

# Générer des chaînes de test
def generer_chaine(taille):
    return 'a' * taille

# Mesurer le temps pour différentes tailles
tailles = range(1000, 10001, 1000)  # De 1000 à 100000 par pas de 1000
algorithmes = [
    ("Reverse", palindrome_reverse),
    ("Two Pointers", palindrome_two_pointers),
    ("Recursion", palindrome_recursion),
    ("Stack", palindrome_stack)
]

resultats = {nom: [] for nom, _ in algorithmes}

for taille in tailles:
    chaine = generer_chaine(taille)
    for nom, algo in algorithmes:
        temps = mesure_temps_execution(algo, chaine)
        resultats[nom].append(temps)

# Tracer le graphe
plt.figure(figsize=(12, 8))
for nom, temps in resultats.items():
    plt.plot(tailles, temps, label=nom, marker='o')

plt.xlabel('Taille de la chaîne')
plt.ylabel('Temps d\'exécution (secondes)')
plt.title('Complexité temporelle des algorithmes de palindrome')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def generate_data(n):
    x = np.arange(1, n+1)
    reverse = x
    two_pointers = np.ones_like(x)
    recursion = x
    stack = x // 2
    return x, reverse, two_pointers, recursion, stack

n = 20
x, reverse, two_pointers, recursion, stack = generate_data(n)

plt.figure(figsize=(12, 8))
plt.plot(x, reverse, label='palindrome_reverse (O(n))', marker='o')
plt.plot(x, two_pointers, label='palindrome_two_pointers (O(1))', marker='s')
plt.plot(x, recursion, label='palindrome_recursion (O(n))', marker='^')
plt.plot(x, stack, label='palindrome_stack (O(n/2))', marker='D')

plt.xlabel('Taille de la chaîne (n)')
plt.ylabel('Espace utilisé')
plt.title('Complexité spatiale des algorithmes de palindrome')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
